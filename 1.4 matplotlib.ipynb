
Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of Matplotlib are:<br>

Generally easy to get started for simple plots <br>
Support for custom labels and texts<br>
Great control of every element in a figure<br>
High-quality output in many formats<br>
Very customizable in general

```python
import matplotlib.pyplot as plt

#the below line makes us coding easy, it lets not to write plt.show() at all

%matplotlib inline
```

```python
# lets create some random variables to use in matplot by using numpy
import numpy as np
x=np.arange(10,40)
y=np.arange(20,50)
print(x,y)
```

```python


##plotting by using plot method  plot 

plt.plot(x,y,'r*',linestyle='dashed')
```

```python
#plotting by using scatter method for easy scatter plot
a= np.random.randn(5,5)
b=np.random.randn(5,5)
plt.scatter(a,b,)
plt.xlabel('Sample x axis')
plt.ylabel('Sample y axis')
plt.title('Sample random scatter plot')
```

```python
## creating subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r*')
plt.subplot(2,2,2)
plt.plot(x,y,'r--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```

```python
#general pi value ;)

np.pi
```

```python
# sine wave using plt
#for x needed to have dividing points at difference of 0.1
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
plt.title('Sine Wave')
```

```python
## subploting sine and cos waves

x= np.arange(0,4*np.pi,0.1)
ysin=np.sin(x)
ycos=np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,ysin,'r')
plt.title('Sine Wave')
plt.subplot(2,1,2)
plt.plot(x,ycos)
plt.title('Cos Wave')
```

```python
#Bar plot#

x = [2,8,10] 
y = [11,16,9]  

plt.plot(x,y)##gives  a random scatter plot
```

```python
plt.bar(x,y)
```

```python
x2 = [3,9,11] 
y2 = [6,15,7] 
plt.bar(x2,y2,color='g')
```

```python
## combining is easy just plot two plots on a single plt object
plt.bar(x,y,color='b')
plt.bar(x2,y2, color='g')
plt.title('bar graph')
```

# Pie Chart

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice

plt.pie(sizes,colors=colors,explode=explode,labels=labels,shadow=False,autopct='%1.1f%%')
plt.axis('equal')
```
