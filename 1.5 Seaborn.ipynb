# JoinPlot
A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot

```python
sns.jointplot(data=tdf,x='tip',y='total_bill')
##here the x, y values are not just naming convetions those are table column names given to interprit in the graph
```

```python
sns.jointplot(data=tdf)
```

```python
sns.jointplot(data=tdf,x='tip',y='total_bill',kind='reg')
```

```python
sns.jointplot(data=tdf, x='tip',y='total_bill', kind='hex')
```
