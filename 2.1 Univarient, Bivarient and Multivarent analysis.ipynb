```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
```

```python
df
```

```python
df['species'].value_counts()
```

```python
df.head()
```

```python
plt.plot(df['sepal_length'],'o')
plt.show()
```

```python
dfsetosa = df.loc[df['species']=='setosa']
## loc funtion works as dataframe.loc[condition]
```

```python
dfsetosa.head()
```

```python
plt.plot(dfsetosa['sepal_length'],'o')
```

```python
dfvergi = df.loc[df['species']=='virginica']
dfversi = df.loc[df['species']=='versicolor']
```

```python
dfvergi.head()
```

```python
plt.plot(dfsetosa['sepal_length'],'o')
plt.plot(dfvergi['sepal_length'],'o')
plt.plot(dfversi['sepal_length'],'o')
plt.show()
```

```python
## to understand the univarient analysis 
plt.plot(dfsetosa['sepal_length'],np.zeros_like(dfsetosa['sepal_length']),'o')
plt.plot(dfvergi['sepal_length'],np.zeros_like(dfsetosa['sepal_length']),'o')
plt.plot(dfversi['sepal_length'],np.zeros_like(dfsetosa['sepal_length']),'o')
plt.xlabel('sepal_length')
plt.show()
##making y axis zero to get good perspection
```

```python
sns.scatterplot(x=df['sepal_length'],y=np.zeros_like(df['sepal_width']),hue=df['species'])
##by using the seaborn the whole process done in one line
```
